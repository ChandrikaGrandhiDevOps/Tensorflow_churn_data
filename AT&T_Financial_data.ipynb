```python
import tensorflow as tf
from tensorflow import keras

# Define the model correctly
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(12,), activation='relu'),  # Fix input_shape
    keras.layers.Dense(1, activation='sigmoid')  # Fix activation function name
])

# Fix compile method spelling and optimizer name
model.compile(optimizer='adam',  # Correct spelling: `compile`, not `complie`
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Assuming x_train and y_train are defined
model.fit(x_train, y_train, epochs=5)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
import seaborn as sn
confusin_matrix = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(confusin_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
